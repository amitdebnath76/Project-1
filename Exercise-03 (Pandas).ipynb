```python
## Gender ( Replace and Convert to integer)
df['Gender'].unique()
```

```python
df['Gender']=df['Gender'].apply(lambda x: x.replace('F','0'))
df['Gender']=df['Gender'].apply(lambda x: x.replace('M','1'))
```

```python
df['Gender']=df['Gender'].astype(int)
```

```python
## Age (Label Encoding and change Data types)
df['Age'].dtype
```

```python
df['Age'].unique()
```

```python
## Import LabelEncoder from Sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Age']=le.fit_transform(df['Age'])
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].astype(int)
```
